# Outliers

```python
df[['balance', 'estimated_salary', 'credit_score']].describe()
```

```python
sns.boxplot(x=df['balance'])
plt.show()
```

```python
sns.boxplot(x='churn', y='balance', data=df)
plt.show()
```

```python
Q1 = df['balance'].quantile(0.25)
Q3 = df['balance'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['balance'] < lower_bound) | (df['balance'] > upper_bound)]
```

```python
outliers
```

```python
outliers[['balance', 'churn']].value_counts()
```

```python
Q1 = df['estimated_salary'].quantile(0.25)
Q3 = df['estimated_salary'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['estimated_salary'] < lower_bound) | (df['estimated_salary'] > upper_bound)]
```

```python
outliers
```

```python
Q1 = df['credit_score'].quantile(0.25)
Q3 = df['credit_score'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['credit_score'] < lower_bound) | (df['credit_score'] > upper_bound)]
```

```python
outliers
```

```python
outliers[['credit_score', 'churn']].value_counts()
```

```python
plt.figure()
plt.plot(df['credit_score'], df['balance'], '.', color='blue')
plt.title('credit_score vs Balance')
plt.xlabel('credit_score')
plt.ylabel('Balance')
plt.show()
```

Outlier analysis was performed on financial variables (balance and estimated salary) to identify extreme customer profiles and assess their relationship with churn. Outliers were analyzed for insight, not removed.
